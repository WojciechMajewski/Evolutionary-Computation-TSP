```python
results = ""
results = r"C:\Users\wojci\Documents\GitHub\Evolutionary-Computation-TSP\LSN_results_optional_LS.txt"
```

```python
with open(results) as f:
    text = f.read()
```

```python
print(text)
```

```python
lines = text.split("\n")
```

```python
lines = [l for l in lines if len(l) > 1]
```

```python
dataset_data = {}
for line in lines:
    if len(line) > 7 and line[:7] == "DATASET":
        dataset_data[line] = []
    else:
        dataset_data[list(dataset_data.keys())[-1]].append(line)
```

```python
dataset_full_data = {}

for key in list(dataset_data.keys()):
    dataset_full_data[key] = {}
    for index, line in enumerate(dataset_data[key]):
        if len(line) > 5 and line[:4] == "Best":
            name = dataset_data[key][index-1]
            single_algorithm_dict = {}
            scores = dataset_data[key][index+1].split(" ")
            single_algorithm_dict["best"] = scores[0]
            single_algorithm_dict["mean"] = scores[1]
            single_algorithm_dict["worst"] = scores[2]
            single_algorithm_dict["best_cycle"] = dataset_data[key][index+2]
            single_algorithm_dict["time"] = dataset_data[key][index+3].split(" ")[1]
            dataset_full_data[key][name] = single_algorithm_dict
```

```python
dataset_names = list(dataset_full_data.keys())
row = dataset_full_data[dataset_names[0]]
alg_names = list(row.keys())
```

```python
for result_type in ["best", "mean", "worst", "time"]:
    print("type:", result_type, "\n")
    print("\t".join(dataset_names))

    for alg in alg_names:
        for dataset in dataset_names:
            val = dataset_full_data[dataset][alg][result_type]
            print(f"{val}", end="\t\t")
        print(alg)
    
    print("\n----------------------------------\n")
```
